```python
import pandas as pd
import numpy as np

IND = pd.read_excel('C:/Users/Satvik Jha/Desktop/Python_test/interval_data.xlsx')
IND.head()
```

```python
melted_df = pd.melt(IND, id_vars=['MPAN', 'Date'], var_name='Hour', value_name='Value')
melted_df.head()
```

```python
melted_df['Week'] = melted_df['Date'].dt.isocalendar().week
result = melted_df.groupby(['MPAN','Week']).agg({'Value':['mean','max','min']}).reset_index()
result.columns = ['MPAN', 'Week', 'Mean', 'Max', 'Min']
result.head()
```

```python
import os

folder_name = 'results2'
if not os.path.exists(folder_name):
    os.makedirs(folder_name)

for mpan in result['MPAN'].unique():
    mpan_df = result[result['MPAN'] == mpan]
    
    file_name = f'{mpan}.csv'
    
    file_path = os.path.join(folder_name, file_name)
    mpan_df.to_csv(file_path, index=False)

    print(f"File '{file_name}' saved successfully.")
```
